THỐNG KÊ DỮ LIỆU ĐỊNH DẠNG CHỮ 

```python
categories = ['job', 'marital', 'education', 'default', 'housing', 'loan', 'contact', 'month','poutcome']
# Tạo 3 hàng biểu đồ mỗi hàng chứa 3 biểu đồ 
fig, axs = plt.subplots(3, 3, sharex=False, sharey=False, figsize=(20, 20))

counter = 0
for category in categories:
    # đếm số lượng người đăng kí và không đăng kí của mỗi loại
    value_counts = data[category].value_counts()
    #print(value_counts)
    trace_x = counter//3
    trace_y = counter%3
    #print(trace_x,trace_y)
    x_pos = np.arange(0, len(value_counts))
    #print(x_pos)
    axs[trace_x, trace_y].bar(x_pos, value_counts.values, tick_label = value_counts.index)
    # Đặt tên cho biểu đồ 
    axs[trace_x, trace_y].set_title(category)
    if category == 'job':
     for tick in axs[trace_x, trace_y].get_xticklabels():
         tick.set_rotation(90)
    
    counter += 1

plt.show()
```

Qua các biểu đồ thống kê số lượng thì ngân hàng có nhiều khách hàng :

 -  làm công việc quản lí và nhân viên văn phòng 
 -  đã kết hôn 
 -  đã tốt nghiệp cấp 3

HÀM TÍNH SỐ LƯỢNG VÀ  VẼ BIỂU ĐỒ 

```python
def calc_draw(x,data):
  d = pd.DataFrame()
  d['yes'] = data[data['y'] == 'yes'][x].value_counts()
  d['no'] = data[data['y'] == 'no'][x].value_counts()
  return d
```

THỐNG KÊ DỮ LIỆU ĐỘ TUỔI, TÌNH TRẠNG KẾT HÔN VỚI KHẢ NĂNG ĐĂNG KÍ 

```python
yes='yes'
no ='no'
fig, axes = plt.subplots(nrows=1, ncols=3,figsize=(15,5))
married = data[data['marital'] == 'married']
single = data[data['marital'] == 'single']
divorced = data[data['marital'] == 'divorced']

ax = sns.histplot(married[married['y']=='yes'].age.dropna(), bins=18, label=yes, ax=axes[0], kde=False, color='g')
ax = sns.histplot(married[married['y']=='no'].age.dropna(), bins=40, label=no, ax=axes[0], kde=False, color='y')
ax.legend()
ax.set_title('married')

ax = sns.histplot(single[single['y']=='yes'].age.dropna(), bins=18, label=yes, ax=axes[1], kde=False, color='g')
ax = sns.histplot(single[single['y']=='no'].age.dropna(), bins=40, label=no, ax=axes[1], kde=False, color='y')
ax.legend()
ax.set_title('single')

ax = sns.histplot(divorced[divorced['y']=='yes'].age.dropna(), bins=18, label=yes, ax=axes[2], kde=False, color='g')
ax = sns.histplot(divorced[divorced['y']=='no'].age.dropna(), bins=40, label=no, ax=axes[2], kde=False, color='y')
ax.legend()
_ = ax.set_title('divorced')
```

```python
for dataset in [data]:
  dataset['age'] = dataset['age'].astype(int)
  dataset.loc[ dataset['age'] <= 11, 'age'] = 0
  dataset.loc[(dataset['age'] > 11) & (dataset['age'] <= 18), 'age'] = 1 
  dataset.loc[(dataset['age'] > 18) & (dataset['age'] <= 22), 'age'] = 2
  dataset.loc[(dataset['age'] > 22) & (dataset['age'] <= 27), 'age'] = 3
  dataset.loc[(dataset['age'] > 27) & (dataset['age'] <= 33), 'age'] = 4
  dataset.loc[(dataset['age'] > 33) & (dataset['age'] <= 40), 'age'] = 5
  dataset.loc[(dataset['age'] > 40) & (dataset['age'] <= 66), 'age'] = 6
  dataset.loc[ dataset['age'] > 66, 'age'] = 7
data['age'].head(5)
```

CHUYỂN ĐỔI CÁC THUỘC TÍNH THÀNH SỐ 

```python
def change(x,y,data):
    for dataset in [data]:
        dataset[x] = dataset[x].map(y)
```

```python
subcribe = {"no": 0 , "yes":1}
change('y',subcribe,data)
# for dataset in [data]:
#   dataset['y'] = dataset['y'].map(subcribe)
data['y'].value_counts()
```

```python
credit = {"no": 0 , "yes":1}
change('default',credit,data)
data['default'].value_counts()
```

```python
house ={"no": 0,"yes": 1}
change('housing',house,data)
data['housing'].value_counts()
```

```python
loan ={"no": 0,"yes": 1}
change('loan',loan,data)
data['loan'].value_counts()
```

```python
contact ={"telephone": 1, "unknown": 0,"cellular": 1}
change('contact',contact,data)
data['contact'].value_counts()
```

```python
poutcome ={"success": 1, "unknown": 0,"failure": 0, "other":0}
change('poutcome',poutcome,data)
data['poutcome'].value_counts()
```

```python
def encode(data, col):
     return pd.concat([data, pd.get_dummies(col, prefix=col.name)], axis=1)
#data = encode(data, data.job)
jobs={"admin.": 1,"unknown": 0,"unemployed": 0,"management": 1,"housemaid": 1,"entrepreneur": 1,"student": 0,"blue-collar": 1,"self-employed": 1,"retired": 0,"technician": 1,"services": 1}
change('job',jobs,data)
data['job'].value_counts()
```
