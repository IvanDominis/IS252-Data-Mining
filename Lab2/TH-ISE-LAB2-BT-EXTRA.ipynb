Đọc dữ liệu và chuyển dữ liệu về dạng list trong list 

```python
path = '/content/drive/MyDrive/Data for ise lab/groceries.csv'
data = dict()
with open(path) as f:
    filedata = csv.reader(f, delimiter=',')
    count = 0
    for line in filedata:
      data[count] = list(set(line))
      count += 1
for i in range(5):
  print(data[i])
```

Hàm tính độ phổ biến 

```python
def frequence(itemset, data, check=False):
    items_count = dict()
    for i in itemset:
        temp_i = {i}
        if check:
            temp_i = set(i)
        for j in data.items():
            if temp_i.issubset(set(j[1])):
                if i in items_count:
                    items_count[i] += 1
                else:
                    items_count[i] = 1
    return items_count
```

Hàm nhóm thuộc tính và độ phổ biến

```python
def support(items_count, data):
    support = dict()
    total = len(data)
    for i in items_count:
        support[i] = items_count[i]/total
    return support
```

```python
min_support = 0.01
```

```python
length = [len(i) for i in data.values()]
itemset = set()
items_chosen = list()
for i in data.values():
        for j in i:
            itemset.add(j)
```

```python
items_count = frequence(itemset, data)
v = support(items_count,data)
for i in v.items():
  print(i[0] ,':',i[1])
```

Tìm tập phổ biến 1 ứng viên 

```python
for i in support(items_count,data).items():
  if i[1] > min_support:
    items_chosen.append(i[0] + ' : ' + str(i[1]))
```
